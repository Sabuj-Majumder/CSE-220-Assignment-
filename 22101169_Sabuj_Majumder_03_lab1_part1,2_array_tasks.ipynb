```python
import math
ar=[10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]

# Mean of a series of numbers stored in an array:

def mean_cal(ar):
  s=0
  n=len(ar)
  for i in ar:
    s+=i
  mean=(s/n)
  return mean

 # Standard deviation of Array:

def st_dv(ar):
  a=0
  b=mean_cal(ar)
  for i in ar:
    a+=(i-b)**2
  sd=math.sqrt(a/(len(ar)-1))
  return round(sd,2)

# A new array where only store the numbers that are at least 1.5 standard deviations away from the mean :

def fun(ar):
  a= mean_cal(ar)
  b= st_dv(ar)
  upr=a+(1.5*b)
  lwr=a-(1.5*b)
  c=0
  for i in ar:
    if lwr>i or i>upr:
      c+=1
  if c==0:
    return None
  else:

    ar2=[None]*c
    j=0
    for i in ar:
      if lwr>i or i>upr:
        ar2[j]=i
        j+=1
    return ar2


print(mean_cal(ar))
print(st_dv(ar))
print(fun(ar))
```
